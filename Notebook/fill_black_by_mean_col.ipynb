```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.python.keras.preprocessing.image import load_img, img_to_array, array_to_img, ImageDataGenerator
from PIL import Image
import json
import os
import math
import random
import pandas as pd
import glob
```

```python
Lslide_dir=r'C:\Users\icech\Desktop\lab2019\2019_4\Data\slide_dis\Modify\Left_slide\splits2'
aloe=r'\007_Aloe_007.png'
lr=load_img(Lslide_dir+aloe)
```

```python
r=img_to_array(lr).astype(np.int16)
br=np.where(r>0,255,r)
```

```python
array_to_img(br)
```

```python
R=r[:,:,0]
G=r[:,:,1]
B=r[:,:,2]
Rmean=np.mean(R[R>0]).astype(np.uint8)
Gmean=np.mean(G[G>0]).astype(np.uint8)
Bmean=np.mean(B[B>0]).astype(np.uint8)
```

```python
fill_r=np.where(r==[0,0,0],[Rmean,Gmean,Bmean],r)
array_to_img(fill_r)
```
